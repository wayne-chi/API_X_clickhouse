```python
import requests
from requests.auth import HTTPBasicAuth

# Replace with your actual key_id and key_secret
key_id = os.getenv('KEY_ID') 
key_secret = os.getenv('KEY_Secret') 

# Replace with the actual value for the town filter
params = {
    "postcode1_filter": '',  # or a specific value
    "postcode2_filter": '',  # or a specific value
    "start_date_filter": '2020-06-06',  # or a specific date
    "end_date_filter": (datetime.today().strftime('%Y-%m-%d') ),    # or a specific date
    "property_type_filter": '',  # or a specific property type
    # "is_new_filter": None,      # or a specific value (0 or 1)
    "town_filter": '',        # or a specific town name
    "county_filter": '',      # or a specific county name
    "skip": 0,          # or a specific offset value
    "limit":3000
}


# Endpoint URL
url = 'https://console-api.clickhouse.cloud/.api/query-endpoints/a21aabf9-37ff-43e7-8121-49eddc351222/run?format=CSVWithNames' #JSONEachRow

# Payload with the query variable
payload = {
    "queryVariables": params
}
```

```python

# Make the POST request with basic authentication and JSON data
response = requests.post(
    url,
    auth=HTTPBasicAuth(key_id, key_secret),
    json=payload,
    headers={"Content-Type": "application/json"}
)

# Check response status and print the result
if response.status_code == 200:
    print("Response Text:", response.content)
else:
    print(f"Error: {response.status_code}, {response.text}")
```

```python
# Use StringIO to read the CSV string
csv_data = StringIO(response.text)

# Read the CSV data into a DataFrame
df = pd.read_csv(csv_data)
df
```
